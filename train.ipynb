```python
import os, fnmatch
import Record_audio as ra
dir = '/Sandbox/AI-ML/AIML-Labs/Hackathon I/speech_data/'
listOfFiles = os.listdir(dir) 
```

```python
import numpy as np
import pandas as pd 
featureVector = []
for audio_file in listOfFiles: 
    feature = ra.get_features(dir+audio_file, 8000,30,128,15)
    #feature.append(audio_file[0:1])
    #print(feature)
    label=audio_file.find("_")
    labelfinal = audio_file[0:label]
    print(labelfinal)
    feature = np.append(feature,labelfinal)
    featureVector.append(feature)
```

```python
df = pd.DataFrame(featureVector)
df.to_csv("file_path.csv")
```

```python
dfNew = pd.read_csv("file_path.csv", header=None)
print(dfNew.shape[1] - 1)
```

```python
#Train, Test split
from sklearn.model_selection import train_test_split
#train, test = train_test_split(dfNew, test_size=0.2)
#train.shape
x=dfNew.values[:, 1:51]
y=dfNew.values[:, -1]
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=.2)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn import tree
from sklearn.model_selection import cross_val_score
from sklearn.neighbors import KNeighborsClassifier

# myList = list(range(1,50))
# neighbors = filter(lambda x: x % 2 != 0, myList)
# cv_scores = []
# for k in neighbors:
#     knn = KNeighborsClassifier(n_neighbors=k)
#     scores = cross_val_score(knn, x_train, y_train, cv=10, scoring='accuracy')
#     cv_scores.append(scores.mean())
#     MSE = [1 - x for x in cv_scores]

#     # determining best k
# optimal_k = neighbors[MSE.index(min(MSE))]
# print ("The optimal number of neighbors is %d" % optimal_k)

clrKN = KNeighborsClassifier()
clrKN = clrKN.fit(x_train, y_train)
outKN = clrKN.predict(x_test)
```

```python
# Prediction accuracy
print("Accuracy for KNeighbors Classifier: " + str(accuracy_score(y_test, outKN)*100)+"%")
```
